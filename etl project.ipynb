```python
employment = pd.read_csv("Unemployment/Unemployment.csv")
employment.head(10)
```

```python
states = ["Alabama","Alaska","Arizona","Arkansas","California","Colorado",
  "Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois",
  "Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland",
  "Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana",
  "Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York",
  "North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania",
  "Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah",
  "Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]
```

```python

employment = employment.loc[employment['Area_name'].isin(states)]
```

```python
employment
```

```python
employment = employment[employment["State"] != "United States"]
```

```python

employment = employment.drop(["State"], axis = 1)
employment
```

```python
employment.rename(columns = {"Area_name":"State"}, inplace = True)
employment
```

```python

employment = employment.drop(['FIPS Code','Rural_urban_continuum_code_2013',
       'Urban_influence_code_2013', 'Metro_2015', 'Civilian_labor_force_2007',
       'Employed_2007', 'Unemployed_2007', 'Unemployment_rate_2007',
       'Civilian_labor_force_2008', 'Employed_2008', 'Unemployed_2008',
       'Unemployment_rate_2008', 'Civilian_labor_force_2009', 'Employed_2009',
       'Unemployed_2009', 'Unemployment_rate_2009'], axis = 1)
employment.head(10)
```

```python
new_employment = employment[["State","Unemployment_rate_2010","Unemployment_rate_2015"]]
new_employment
```

```python
new_df = pd.merge(new_employment, new_insurance, how='left', sort=True, on=["State"])

new_df
```

```python
new_df.dropna()
```

```python
connection_string = "postgres:postgres@localhost:5432/Etl_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
new_insurance.to_sql(name='new_insurance', con=engine, if_exists='append', index=True)
```

```python
new_employment.to_sql(name="new_employment", con=engine, if_exists='append', index=True)
```
